```python
numeric_cols = df.select_dtypes(include = ["int64", "float64"]).columns

for col in numeric_cols:
    plt.figure(figsize = (8,5))
    sns.histplot(df[col], kde = True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

### ***Distribution Summary:***

*The following table summarizes the distribution shapes and notes for each numeric column based on the histograms:*


| Column               | Shape          | Note |
|----------------------|---------------|------|
| Unit price           | Irregular     | Distribution without a clear pattern; most values are around or above 80. |
| Quantity             | Skewed        | A gap (empty bar) in the middle of the plot; bars are asymmetrical and most values are above 80. |
| Customer type        | Binary        | Only two bars: one on the left (near 0) and one on the right (near 1); the left bar is slightly taller. |
| Tax 5%               | Slight skew   | Left bar taller than the right; the first bar is shorter than the second. |
| Sales                | Slight skew   | Similar to Tax 5%; left bar taller with a small difference between bars. |
| cogs                 | Slight skew   | Similar to Tax 5%; left bar taller with a small difference between bars. |
| gross income         | Slight skew   | Similar to Tax 5%; left bar taller with a small difference between bars. |
| Rating               | Irregular     | Large variation in bar heights; most values are close to or above 80. |


```python
plt.figure(figsize = (10,8))
corr_matrix = df.corr(numeric_only = True)
sns.heatmap(corr_matrix, annot = True, cmap = "coolwarm", fmt = ".2f")
plt.title("Correlation Matrix")
plt.show()
```

### ***Correlation Matrix Analysis***

*- Focus on the highest and lowest correlation values.*

*- A correlation close to **+1** means a strong positive relationship (when one increases, the other tends to increase as well).*

*- A correlation close to **-1** means a strong negative ralationship (when one increases, the other tends to decrease).*

*- Acorrelation close to **0** means there is no significant linear relationship.*

*- In this dataset, as expected in sales data, columns like `Tax 5%`, `Sales`, `cogs`, and `gross income` show very high positive correlations with each other.*

```python
plt.figure(figsize = (8,6))
sns.boxplot(data = df, x = "Gender", y = "Sales")
plt.title("Sales by Gender")
plt.show()

plt.figure(figsize = (8,6))
sns.boxplot(data = df, x = "Customer type", y = "Sales")
plt.title("Sales by Customer Type")
plt.show()
```

### ***Sales Comparison by Gender and Customer Type***

*- **Gender**: The average sales for **female** customers are slighly higher than for male customers, with a few high-value outliers in the male group.*

*- **Customer Type**: Encoded as `0` and `1`, where values show a similar pattern to gender. Customers labeled as `0` have slightly higher average sales then those labeled as `1`.*

```python
df.groupby("Gender")["Sales"].mean()
df.groupby("Customer type")["Sales"].mean()
df.groupby("Product line")["Sales"].mean().sort_values(ascending= False)
```
