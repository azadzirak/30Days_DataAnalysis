## **Day 29 of 30 days Data Analysis**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:\\Users\\LENOVO\\Documents\\Programming\\30days_data_analysis\\week02_sql\\supermarket_clean_d08.csv")
```

```python
import matplotlib.pyplot as plt

# 2x2 dashboard layout 
fig, axes = plt.subplots(2, 2, figsize=(12,8))

# 1. Sales by City 
city_sales = df.groupby("City")["Sales"].sum().sort_values(ascending=False)
city_sales.plot(kind="bar", ax=axes[0,0], color="blue")
axes[0,0].set_title("Sales by City")
axes[0,0].set_ylabel("Total Sales")

# 2. Monthly Sales Trend
if "Date" in df.columns and not np.issubdtype(df["Date"].dtype, np.datetime64):
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["MonthName"] = df["Date"].dt.month_name()
monthly_sales = df.groupby("Month")["Sales"].sum()
monthly_sales.plot(kind="line", marker="*", ax=axes[0,1], color="purple")
axes[0,1].set_title("Monthly Sales Trend")
axes[0,1].set_ylabel("Total Sales")

# 3. Payment Methods Distribution
payment_counts = df["Payment"].value_counts()
axes[1,0].pie(payment_counts.values, labels=payment_counts.index,
              autopct="1%.1f%%", startangle=90, colors=["teal", "skyblue", "green"])
axes[1,0].set_title("Payment Methods")

# 4. Rating Distribution
df["Rating"].plot(kind="hist", bins=10, ax=axes[1,1], color="red", alpha=0.7)
axes[1,1].set_title("Customer Ratings")
axes[1,1].set_xlabel("Rating")

plt.tight_layout()
plt.savefig("day29_dashboard.png")
plt.show()
```
