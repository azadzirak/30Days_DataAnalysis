## **Day 30 of 30 days Data Analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load final dataset
df = pd.read_csv("C:\\Users\\LENOVO\\Documents\\Programming\\30days_data_analysis\\week02_sql\\supermarket_clean_d08.csv")

# === KPIs ===
total_sales = df["Sales"].sum()
avg_rating = df["Rating"].mean()
top_city = df.groupby("City")["Sales"].sum().idxmax()
top_payment = df["Payment"].mode()[0]

# === Dashboard ===
fig, axes = plt.subplots(2, 2, figsize=(14,10))

# KPI Tiles (just text)
axes[0,0].axis("off")
axes[0,0].text(0.1, 0.8, f"Total Sales:\n${total_sales:,.0f}", fontsize=16, weight="bold")
axes[0,0].text(0.1, 0.55, f"Top City:\n{top_city}", fontsize=16, weight="bold")
axes[0,0].text(0.1, 0.3, f"Top Payment:\n{top_payment}", fontsize=16, weight="bold")
axes[0,0].text(0.1, 0.05, f"Avg Rating:\n{avg_rating:.2f}", fontsize=16, weight="bold")

# Sales by City
city_sales = df.groupby("City")["Sales"].sum().sort_values(ascending=False)
sns.barplot(x=city_sales.index, y=city_sales.values, ax=axes[0,1], palette="crest")
axes[0,1].set_title("Sales by City")
axes[0,1].set_ylabel("Total Sales")

# Monthly Sales Trend
df["Date"] = pd.to_datetime(df["Date"])
monthly_sales = df.groupby(df["Date"].dt.to_period("M"))["Sales"].sum()
monthly_sales.index = monthly_sales.index.astype(str)
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker="*", ax=axes[1,0], color="purple")
axes[1,0].set_title("Monthly Sales Trend")
axes[1,0].set_ylabel("Sales")

# Payment Distribution
payment_counts = df["Payment"].value_counts()
axes[1,1].pie(payment_counts.values, labels=payment_counts.index, autopct="%1.1f%%",
              colors=sns.color_palette("pastel"))
axes[1,1].set_title("Payment Methods")

plt.suptitle("Final Dashboard - Day 30", fontsize=18, weight="bold")
plt.tight_layout()
plt.savefig("day30_dashboard.png")
plt.show()
```
