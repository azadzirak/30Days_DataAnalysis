### ***The value_counts() method in Pandas is used to count the occurrences of unique values within a Series (a single column of a DataFrame). It returns a new Series where the index represents the unique values from the original Series, and the values represent their respective counts.***

```python
df.groupby("Product line")["Sales"].mean().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
df["Gender"].value_counts().plot(kind="bar", color="darkblue")
plt.title("Distribution By Gender")
plt.show()
```

```python
df["City"].value_counts().plot(kind= "pie", autopct= "%1.1f%%", startangle= 90, shadow= True)
plt.title("Customer Distribution By City")
plt.ylabel("")
plt.show()
```

```python
df["Sales"].plot(kind= "hist", bins= 20, color= "green", edgecolor= "black")
plt.title("Distribution Of Total Purchase Amount")
plt.xlabel("Total Amount")
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize= (10,6))
sns.boxplot(x= "Product line", y= "Sales", data= df)
plt.title("Total Purchase By Product Line")
plt.xticks(rotation= 45)
plt.show()
```
