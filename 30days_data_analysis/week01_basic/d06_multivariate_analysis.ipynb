### ***Explore relationship between multiple features such as Gender, Product line, Sales, and Ratings.***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df = pd.read_csv("C:\\Users\\LENOVO\\Documents\\Programming\\30days_data_analysis\\datasets\\SuperMarket Analysis.csv")
df.head()
```

### ***Product Line vs. Total Sales by Gender***

```python
plt.figure(figsize= (10,6))
sns.boxplot(data= df, x= "Product line", y= "Sales", hue= "Gender")
plt.title("Sales distribution by Product line and Gender")
plt.xticks(rotation= 45)
plt.show()
```

### ***Insight:***

*- In all categories except *Health and beauty* women spend more.*

*- **Food and beverages** category has the highest sales and **Fashion accessories** category has the lowest sales.*

### ***Product Line vs. Rating***

```python
plt.figure(figsize= (10,6))
sns.boxplot(data= df, x= "Product line", y= "Rating", palette= "coolwarm")
plt.xticks(rotation= 45)
plt.title("Customer Ratings per Product Line")
plt.show()
```

### ***Insight:***

*- Average ratings across product lines are relatively close, with **Food and beverages** slightly higher and **Health and beauty** showing more variance.*

### ***Pivot Table: Gender x City***

```python
pivot= pd.pivot_table(df, index= "Gender", columns= "City", values= "Sales", aggfunc= "sum")
pivot
```

### ***Observation:***

*- In all cities, women made more purchases and contributed a higher total sales amount.*

### ***Count of Purchases by Gender per Product***

```python
plt.figure(figsize= (10,6))
sns.countplot(data= df, x= "Product line", hue= "Gender")
plt.xticks(rotation= 45)
plt.title("Gender distribution per Product line")
plt.show()
```

### ***Insight:***

*- The amount of purchases by women was higher than that by men in all categories.*

*- The highest sales volume was in the category of **Fashion accessories**.*

### ***Summary:***

*- Fashion accessories had the highest spending, especially among female customers.*

*- Food and beverages received the highest average customer ratings.*

*- Health and beauty products showed interesting gender distribution, with a noticeable preference gap.*

*- Across all cities, women made more purchases and generated higher total sales than men.*
