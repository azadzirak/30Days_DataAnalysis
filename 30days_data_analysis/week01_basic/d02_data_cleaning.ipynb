## **Day 02 of 30 days Data Analysis**

### ***Grouped analysis and visualization***

```python
import pandas as pd
df = pd.read_csv("C:\\Users\\LENOVO\\Documents\\Programming\\30days_data_analysis\\datasets\\SuperMarket Analysis.csv")
df.head()
```

```python
df.groupby("City")["Sales"].sum().sort_values(ascending= False)
```

### ***Insight: City Sales***

*- The city with the highest total sales is **Naypyitaw**, followed by **Yangon** and **Mandalay**.*

*- This suggests that Naypyitaw is a key market and might deserve more targeted marketing or inventory.*

```python
city_sales = df.groupby("City")["Sales"].sum().sort_values(ascending= False)
import matplotlib.pyplot as plt
city_sales.plot(kind= "bar", color= "blue")
plt.title("Total Sales By City")
plt.ylabel("Total Sales")
plt.xlabel("City")
plt.xticks(rotation= 0)
plt.show()
```

### ***City-Based Sales Analysis***

*From the grouped sales data, we observe that:*

*- **Naypyitaw** has the highest total sales.*
*- The difference between cities is not huge, but still noticeable.*
*- It may be worth exploring what factors contribute to higher sales in Naypyitaw.*

*→ Suggestion: Investigate customer type or product preferences in Naypyitaw.*


```python
df.groupby("Product line")["Sales"].mean().sort_values(ascending= False)
```

### ***Insight: Product line Sales***

*- The product line with the highest total sales is **Home and lifestyle**, followed by **Sports and travel** and **Health and beauty**.*

*- This suggests that Home and lifestyle is a key market and might deserve more targeted marketing or inventory.*

```python
product_line_sales = df.groupby("Product line")["Sales"].sum().sort_values(ascending= False)
product_line_sales.plot(kind= "bar", color= "purple")
plt.title("Total Sales By Product Line")
plt.ylabel("Total Sales")
plt.xlabel("Product Line")
plt.xticks(rotation= 45)
plt.show()
```

### ***Product-line-Based Sales Analysis***

*From the grouped sales data, we observe that:*

*- **Home and lifestyle** has the highest total sales.*
*- It may be worth exploring what factors contribute to higher sales in Home and lifestyle.*

*→ Suggestion: Investigate customer type or product preferences in Home and lifestyle product line.*

```python
df["Payment"].value_counts()
```

### ***Insight: Pyment Sales***

*- The Payment with the highest total sales is **Ewallet**, followed by **Cash** and **Credit card**.*

*- This suggests that Ewallet is a key market and might deserve more targeted marketing or inventory.*

```python
payment_sales = df.groupby("Payment")["Sales"].sum().sort_values(ascending= False)

payment_sales.plot(kind= "bar", color= "green")
plt.title("Total Sales By Payment")
plt.ylabel("Total Sales")
plt.xlabel("Payment", loc= 'center')
plt.xticks(rotation= 45)
plt.show()
```

## ***Payment-Based Sales Analysis***

*From the grouped payment based sales data, we observe that:*

*- **Cash** has the highest total payment method.*

*- **Ewallet** has the second highest payment method overall.*

*- **Credit Card** has the lowest payment method overall.*

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(x="Product line", y="Sales", data=df, estimator="mean")
plt.title("Average Sales by Product Line")
plt.xticks(rotation=45)
plt.show()
```

*- **Home and lifestyle** has the highest average sales by Product line.*

```python
sns.countplot(x="City", data=df)
plt.title("Number of Transactions by City")
plt.show()
```

*- **Yangon** has the highest Number of transaction by city.*

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="Blues")
plt.title("Correlation Matrix")
plt.show()
```
