## **Day 24 of 30 days Data Analysis**

### ***Creating simple dashboard with matplotlib***

### ***1. Load dataset***

```python
import pandas as pd

df = pd.read_csv("C:\\Users\\LENOVO\\Documents\\Programming\\30days_data_analysis\\week02_sql\\supermarket_clean_d08.csv")
df.info()
df.head()
```

### ***2. Creat final dashboard***

```python
import matplotlib.pyplot as plt

# Step one
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month
monthly_sales = df.groupby("Month")["Sales"].sum()
branch_sales = df.groupby("Branch")["Sales"].sum()
payment_counts = df["Payment"].value_counts()
ratings = df["Rating"]

# Step two
fig, axes = plt.subplots(2, 2, figsize=(12,8))

# Monthly Sales
monthly_sales.plot(kind="line", marker="*", color="red", ax=axes[0,0])
axes[0,0].set_title("Monthly Sales Trend")

# Branch Sales
branch_sales.plot(kind="bar", color="green", ax=axes[0,1])
axes[0,1].set_title("Sales by Branch")

# Payment Distribution 
axes[1,0].pie(payment_counts.values, labels=payment_counts.index, autopct="%1.1f%%", colors=["purple", "skyblue", "pink"])
axes[1,0].set_title("Payment Methods")

# Ratings
ratings.plot(kind="hist", bins=10, color="gray", alpha=0.7, ax=axes[1,1])
axes[1,1].set_title("Ratings Distribution")

plt.tight_layout()
plt.show()
```

```python
fig.savefig("dashboard_summary.png")
```

### ***Final Dashboard Insights - Day 24***

*- **Sales Trends:** Peak sales observed in February and on weekends.*

*- **Branch Performance:** Branch A leads in total revenue.*

*- **Payment Methods:** E-wallet is the most used payment method.*

*- **Customer Ratings:** Average rating ~7.0 with positive skew, indicating mostly good feedback.*
